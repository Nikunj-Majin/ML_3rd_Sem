```python
import numpy as np
import matplotlib.pyplot as plt
import random

x1=np.array([[random.uniform(0,2),random.uniform(0,2)] for i in range(100)])
x2=np.array([[random.uniform(2,4),random.uniform(2,4)] for i in range(100)])

plt.scatter([i[0] for i in x1],[i[1] for i in x1])
plt.scatter([i[0] for i in x2],[i[1] for i in x2])
```

```python
x_1 = np.mean(x1,axis = 0)
x_2 = np.mean(x2,axis = 0)

print(x_1)
print(x_2)
```

```python
plt.scatter(x1[:,0],x1[:,1])
plt.scatter(x2[:,0],x2[:,1])

plt.scatter(x_1[0],x_1[1])
plt.scatter(x_2[0],x_2[1])
plt.show()
```

```python
def test(a):
    test_pt = np.array(a)
    dist1 = np.linalg.norm(a-x_1)
    dist2 = np.linalg.norm(a-x_2)
    if dist1 < dist2:
        print("It belongs to Cluster1")
    else:
        print("It belongs to Cluster2")
        
test([2,2])
```
